```python
data_victimas = pd.read_excel("homicidios.xlsx", sheet_name="VICTIMAS")
```

A continuación vamos a segmentar los datos delos siniestros en diferentes dataframes que luego exportaremos a la carpeta "data"

```python
data_hechos.columns
```

```python
incidentes = data_hechos[['ID', 'N_VICTIMAS', 'FECHA', 'HORA', 'PARTICIPANTES', 'VICTIMA', 'ACUSADO']]
incidentes.head()
```

```python
localizacion = data_hechos[['ID', 'FECHA', 'AAAA', 'MM', 'DD', 'HORA', 'HH',
       'LUGAR_DEL_HECHO', 'TIPO_DE_CALLE', 'Calle', 'Altura', 'Cruce',
       'Dirección Normalizada', 'COMUNA', 'XY (CABA)', 'pos x', 'pos y']]
localizacion.head()
```

```python
tipos_de_calles = {}
for i in data_hechos['TIPO_DE_CALLE'].unique():
    tipos_de_calles[f"{i}"] = ""


tipos_de_calles["CALLE"] = "Arteria cuya calzada tiene un ancho conprendido entre cinco (5) y trece (13) metros. Incluye paisaje."
tipos_de_calles["AVENIDA"] = "Arteria cuya calzada tiene un ancho total de por lo menos trece (13) metros."
tipos_de_calles["AUTOPISTA"] = "Via multicarril con calzadas para ambas manos separadas fisicamente, sin creces de nivel, con accesos controlados y sin ingreso desde los predios frentistas lindates. Incluye ingresos y egresos de las mismas y distribuciones."
tipos_de_calles["GRAL PAZ"] = "Avenida General Paz, ambos sentidos. Incluye ingresos y egresos de las mismas, distribuidores y calle colectora."
```

```python
df_calles = pd.DataFrame(tipos_de_calles, index= ['Descripcion'])
df_calles = df_calles.T.reset_index()
df_calles = df_calles.reset_index()
df_calles.rename(columns= {'index': 'Calificacion', 'level_0': 'ID_calle'}, inplace= True)
df_calles
```

```python
for n, txt in df_calles[['ID_calle','Calificacion']].values:
    localizacion['TIPO_DE_CALLE'] = localizacion['TIPO_DE_CALLE'].str.replace(f"{txt}", f"{n}")
```

```python
victim_class = {}
for e in incidentes['VICTIMA'].unique():
    victim_class[f"{e}"] = input(f"Descripción de: {e}")
```

```python
victim_df = pd.DataFrame(victim_class, index= ['Descripcion'])
```

```python
victim_df = victim_df.T.reset_index()
victim_df = victim_df.reset_index()
victim_df.rename(columns= {"index": "Calificacion", "level_0": "Type_Victim"}, inplace= True)
victim_df
```

```python
for ix, txt in victim_df[['Type_Victim', 'Calificacion']].values:
    incidentes['VICTIMA'] = incidentes['VICTIMA'].str.replace(f"{txt}", f"{ix}")

incidentes
```

```python
acusado_class = {}
for x in incidentes['ACUSADO'].unique():
    acusado_class[f"{x}"] = input(f"{x}")
```

```python
acusado_df = pd.DataFrame(acusado_class, index=["Descripción"])
acusado_df = acusado_df.T.reset_index().reset_index()
acusado_df.rename(columns= {"index": "Calificacion", "level_0": "id_Acusado"}, inplace= True)
acusado_df
```

```python
for z, txt in acusado_df[['id_Acusado', 'Calificacion']].values:
    incidentes['ACUSADO'] = incidentes['ACUSADO'].str.replace(f"{txt}", f"{z}")

incidentes
```

```python
for i, txt in victim_df[['Type_Victim', 'Calificacion']].values:
    data_victimas['VICTIMA'] = data_victimas['VICTIMA'].str.replace(f"{txt}", f"{i}")
```

```python
incidentes.to_csv("./data/Incidente.csv", index= False)
data_victimas.to_csv("./data/Victimas.csv", index= False)
df_calles.to_csv("./data/tipo_Calle.csv", index= False)
localizacion.to_csv("./data/Localizacion.csv", index= False)
victim_df.to_csv("./data/Type_Victimas.csv", index= False)
acusado_df.to_csv("./data/Type_Acusados.csv", index= False)
```
